```python
from pathlib import Path
import json
import fitz
```

```python
ROOT = Path.home() / "gemini HTML output"

LANGUAGE = "Assamese"
DOC_NAME = "Lentil_KVK , Thoubal ,Manipur"

WORKDIR_ROOT = ROOT / "Workdir" / LANGUAGE / DOC_NAME
FINAL_OUTPUT_DIR = WORKDIR_ROOT / "final_output"
FINAL_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

START_PAGE = 1
END_PAGE = 2

FINAL_PDF_PATH = FINAL_OUTPUT_DIR / f"{DOC_NAME}_translated_pages_{START_PAGE:03d}_to_{END_PAGE:03d}.pdf"

print("Workdir root:", WORKDIR_ROOT)
print("Final output dir:", FINAL_OUTPUT_DIR)
print("Final PDF path:", FINAL_PDF_PATH)
print("Workdir exists:", WORKDIR_ROOT.exists())
```

```python
pdf_paths = []

for page_num in range(START_PAGE, END_PAGE + 1):
    page_name = f"page_{page_num:03d}"
    pdf_path = WORKDIR_ROOT / page_name / "final_page.pdf"
    pdf_paths.append(pdf_path)

print("Total expected PDFs:", len(pdf_paths))
print("First 5:")
for p in pdf_paths[:5]:
    print("  ", p, "->", p.exists())

print("Last 5:")
for p in pdf_paths[-5:]:
    print("  ", p, "->", p.exists())
```

```python
def merge_pdfs(pdf_paths, output_path: Path, strict: bool = True):
    merged_doc = fitz.open()
    missing = []

    for pdf_path in pdf_paths:
        if not pdf_path.exists():
            missing.append(str(pdf_path))
            if strict:
                merged_doc.close()
                raise FileNotFoundError(f"Missing PDF: {pdf_path}")
            continue

        src = fitz.open(pdf_path)
        merged_doc.insert_pdf(src)
        src.close()

    output_path.parent.mkdir(parents=True, exist_ok=True)
    merged_doc.save(str(output_path))
    merged_doc.close()

    return missing
```

```python
missing_files = merge_pdfs(
    pdf_paths=pdf_paths,
    output_path=FINAL_PDF_PATH,
    strict=True,
)

print("Merge complete")
print("Final PDF:", FINAL_PDF_PATH)
print("Missing files:", missing_files)
```
