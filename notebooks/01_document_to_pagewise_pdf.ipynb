```python
from pathlib import Path
import fitz
```

```python
ROOT = Path.home() / "gemini HTML output"

LANGUAGE = "Assamese"
DOC_NAME = "Lentil_KVK , Thoubal ,Manipur"

SOURCE_PDF_PATH = ROOT / "Data" / LANGUAGE / f"{DOC_NAME}.pdf"
WORKDIR_ROOT = ROOT / "Workdir" / LANGUAGE / DOC_NAME

PAGES_ROOT = WORKDIR_ROOT
OVERWRITE_EXISTING = False

print("Source PDF:", SOURCE_PDF_PATH)
print("Workdir root:", WORKDIR_ROOT)
print("PDF exists:", SOURCE_PDF_PATH.exists())
```

```python
def split_pdf_to_page_folders(
    source_pdf_path: Path,
    output_root: Path,
    overwrite_existing: bool = False,
):
    if not source_pdf_path.exists():
        raise FileNotFoundError(f"Source PDF not found: {source_pdf_path}")

    doc = fitz.open(source_pdf_path)
    total_pages = len(doc)

    output_root.mkdir(parents=True, exist_ok=True)
    saved_files = []

    for i in range(total_pages):
        page_num = i + 1
        page_name = f"page_{page_num:03d}"
        page_dir = output_root / page_name
        page_dir.mkdir(parents=True, exist_ok=True)

        out_pdf_path = page_dir / f"{page_name}.pdf"

        if out_pdf_path.exists() and not overwrite_existing:
            saved_files.append(out_pdf_path)
            continue

        new_doc = fitz.open()
        new_doc.insert_pdf(doc, from_page=i, to_page=i)
        new_doc.save(out_pdf_path)
        new_doc.close()

        saved_files.append(out_pdf_path)

    doc.close()
    return saved_files
```

```python
page_pdf_files = split_pdf_to_page_folders(
    source_pdf_path=SOURCE_PDF_PATH,
    output_root=PAGES_ROOT,
    overwrite_existing=OVERWRITE_EXISTING,
)

print(f"Created or found {len(page_pdf_files)} page PDFs")
print("First 5:")
for p in page_pdf_files[:5]:
    print("  ", p)

print("Last 5:")
for p in page_pdf_files[-5:]:
    print("  ", p)
```
