```python
from pathlib import Path
import fitz
```

```python
APP_ROOT = Path("ADD YOUR PROJECT ROOT PATH HERE")

DOC_NAME = "ADD DOCUMENT NAME HERE"
SOURCE_PDF_PATH = APP_ROOT / "input" / "ADD DOCUMENT NAME HERE.pdf"   
START_PAGE = ADD FIRST PAGE NUMBER TO TRANSLATE(EX: 1)   
END_PAGE = ADD LAST PAGE NUMBER TO TRANSLATE(EX: 20)     

JOB_ROOT = APP_ROOT / "workdir" / DOC_NAME
PAGES_PDF_ROOT = JOB_ROOT / "pages_pdf"
PAGES_PDF_ROOT.mkdir(parents=True, exist_ok=True)

print("Source PDF:", SOURCE_PDF_PATH)
print("Output root:", PAGES_PDF_ROOT)
print("PDF exists:", SOURCE_PDF_PATH.exists())
```

```python
def split_pdf_into_pagewise_pdfs(
    source_pdf_path: Path,
    output_root: Path,
    start_page: int,
    end_page: int,
):
    if not source_pdf_path.exists():
        raise FileNotFoundError(f"Source PDF not found: {source_pdf_path}")

    if start_page < 1 or end_page < 1:
        raise ValueError("start_page and end_page must be >= 1")

    if start_page > end_page:
        raise ValueError("start_page cannot be greater than end_page")

    src_doc = fitz.open(source_pdf_path)
    total_pages = len(src_doc)

    if end_page > total_pages:
        raise ValueError(f"PDF has only {total_pages} pages, but end_page={end_page}")

    output_root.mkdir(parents=True, exist_ok=True)
    saved_files = []

    for page_num in range(start_page, end_page + 1):
        page_name = f"page_{page_num}"
        page_dir = output_root / page_name
        page_dir.mkdir(parents=True, exist_ok=True)

        out_pdf_path = page_dir / f"{page_name}.pdf"

        new_doc = fitz.open()
        new_doc.insert_pdf(src_doc, from_page=page_num - 1, to_page=page_num - 1)
        new_doc.save(out_pdf_path)
        new_doc.close()

        saved_files.append(out_pdf_path)

    src_doc.close()
    return saved_files
```

```python
page_pdf_files = split_pdf_into_pagewise_pdfs(
    source_pdf_path=SOURCE_PDF_PATH,
    output_root=PAGES_PDF_ROOT,
    start_page=START_PAGE,
    end_page=END_PAGE,
)

print(f"Generated {len(page_pdf_files)} page-wise PDFs:")
for f in page_pdf_files:
    print(f)
```
