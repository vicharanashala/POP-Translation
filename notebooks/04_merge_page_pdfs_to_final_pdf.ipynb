```python
from pathlib import Path
import fitz
```

```python
APP_ROOT = Path("/home/dhruva/pop_gemini_direct_pdf_pipeline")

DOC_NAME = "kannada_test_job"
JOB_ROOT = APP_ROOT / "workdir" / DOC_NAME
PAGES_PDF_ROOT = JOB_ROOT / "pages_pdf"
FINAL_OUTPUT_DIR = JOB_ROOT / "final_output"
FINAL_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

START_PAGE = 16
END_PAGE = 20

FINAL_PDF_PATH = FINAL_OUTPUT_DIR / f"{DOC_NAME}_translated_pages_{START_PAGE}_to_{END_PAGE}.pdf"

print("Pages root:", PAGES_PDF_ROOT)
print("Final output dir:", FINAL_OUTPUT_DIR)
print("Final PDF path:", FINAL_PDF_PATH)
print("Pages root exists:", PAGES_PDF_ROOT.exists())
```

```python
pdf_paths = []

for page_num in range(START_PAGE, END_PAGE + 1):
    page_name = f"page_{page_num}"
    pdf_path = PAGES_PDF_ROOT / page_name / "translated_en.pdf"
    pdf_paths.append(pdf_path)

for p in pdf_paths:
    print(p, "->", p.exists())
```

```python
def merge_pdfs(pdf_paths, output_path: Path, strict: bool = True):
    merged_doc = fitz.open()
    missing = []

    for pdf_path in pdf_paths:
        if not pdf_path.exists():
            missing.append(str(pdf_path))
            if strict:
                merged_doc.close()
                raise FileNotFoundError(f"Missing PDF: {pdf_path}")
            continue

        src = fitz.open(pdf_path)
        merged_doc.insert_pdf(src)
        src.close()

    output_path.parent.mkdir(parents=True, exist_ok=True)
    merged_doc.save(str(output_path))
    merged_doc.close()

    return missing
```

```python
missing_files = merge_pdfs(
    pdf_paths=pdf_paths,
    output_path=FINAL_PDF_PATH,
    strict=True,
)

print("Merge complete")
print("Final PDF:", FINAL_PDF_PATH)
print("Missing files:", missing_files)
```
